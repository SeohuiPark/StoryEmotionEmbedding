```python
import pandas as pd
import numpy as np
import glob
import collections
from collections import Counter
```

```python
def data_load():
    user_data = pd.read_csv("roc_humanrater.csv")
    question_data = pd.read_csv("total_sample_df_suffle_survey_edited.csv")
    
    return user_data, question_data
```

```python
if __name__ == "__main__":
    user_data, question_data = data_load()
```

```python
user_data_emo = user_data[1::2]
user_data_pos = user_data[0::2]
```

```python
map_dic = {'Anger':0, 'Anticipation':1,'Disgust':2, 'Fear':3, 'Joy':4, 'Trust':5,'Sadness':6,'Surprise':7}
func = lambda x: map_dic.get(x, x)
```

```python
user_data_emo_label  = user_data_emo.copy()
user_data_emo_label['user1'] = user_data_emo.user1.map(func)
user_data_emo_label['user2'] = user_data_emo.user2.map(func)
user_data_emo_label['user3'] = user_data_emo.user3.map(func)
user_data_emo_label['user4'] = user_data_emo.user4.map(func)
```

```python
user_data_emo_label = user_data_emo_label.reset_index(drop = True)
user_data_emo_label['label'] = question_data['emotionlabel']
sort_by_label = user_data_emo_label.sort_values(by = ['label'])
```

```python
total_human_rate_list = []
for i in range(8):
    human_rate_list = list(sort_by_label[sort_by_label['label']==i].iloc[:,1:-1].values.flatten())
    total_human_rate_list.append(human_rate_list)
```
