## 1. 각 data set 에서 랜덤하게 10 문장 뽑기 

```python
def data_load(files):
    all_data = []
    for file in files:
        raw_data = pd.read_csv(file,header=0)
        data = raw_data.iloc[:,2:]
        all_data.append(data)
        
    return all_data
```

```python
def random_sampling(file_path, emo_list, all_data):
    all_sample_list = []
    for idx, data in enumerate(all_data):
        
        # story title 은 샘플에서 제외 
        data = data[data['sequence']!= 'storytitle']
        sample = data.sample(n=10)
        
        # sample csv로 저장하기
        sample.to_csv(file_path+'sample/{0}_emo_sample.csv'.format(emo_list[idx]))
        print(idx, emo_list[idx],'isdone')
        all_sample_list.append(sample)
                
    return all_sample_list 
```

```python
def concat_and_suffle_data(all_sample_list):
    all_sample_df  = pd.concat(all_sample_list)
    all_sample_df = all_sample_df.sample(frac = 1).reset_index(drop = True)
    
    return all_sample_df
```

```python
if __name__ == "__main__":
    emo_list = ['anger','anticipation','disgust','fear','joy', 'trust','sadness','surprise']
    
    file_path = 'C:/Users/Administrator/Anaconda3/envs/temp/EmotionEmbedding/emotion8_finalcode/finalresult/result3/'
    file_name = '*_emo_result.csv'
    files = glob.glob(file_path+file_name)
        
    all_data = data_load(files) # load daa
    all_sample_list = random_sampling(file_path, emo_list, all_data)

    all_sample_df = concat_and_suffle_data(all_sample_list)
    all_sample_df.to_csv(file_path+"sample/all_sample_df_suffle.csv")
```
