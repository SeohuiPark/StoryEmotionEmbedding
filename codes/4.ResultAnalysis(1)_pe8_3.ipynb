# 각 감정 개수 확인 및 문장 뽑기 

```python
def check_emotionlabel(emo_list):
    result_dic = final_df['emotionlabel'].value_counts()
    result_df = pd.DataFrame.from_dict(result_dic)
    result_df = result_df.sort_index()
    
    new = pd.DataFrame()
    new['emotion'] = emo_list
    new['count'] = result_df['emotionlabel']
    new['per'] = [round((i/sum(new['count']))*100,2) for i in new['count']]
    return new
```

# 각 감정별 키워드 랭킹 뽑기 

```python
def get_sentiword_ranking(word_list):
    emo_word_dict = dict(collections.Counter(word_list))
    emo_rank = pd.DataFrame()
    emo_rank['word'] = emo_word_dict.keys()
    emo_rank['counts'] = emo_word_dict.values()
    emo_rank = emo_rank.sort_values(by = 'counts',ascending  = False)
    emo_rank = emo_rank.reset_index(drop = True)
    
    return emo_rank
```
