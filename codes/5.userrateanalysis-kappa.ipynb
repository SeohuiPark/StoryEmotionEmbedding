```python
def data_load():
    user_data = pd.read_csv("roc_humanrater.csv")
    question_data = pd.read_csv("total_sample_df_suffle_survey_edited.csv")
    
    return user_data, question_data
```

```python
map_dic = {'Anger':0, 'Anticipation':1,'Disgust':2, 'Fear':3, 'Joy':4, 'Trust':5,'Sadness':6,'Surprise':7}
func = lambda x: map_dic.get(x, x)
```
